```python
import random
num = random.randrange(0,10) 
guess = int(input("Enter any number from 0 to 10 "))
while num!= guess:
    if guess < num:
        print(" no is low")
        guess = int(input("Enter any number "))
    elif guess > num:
        print (" no is high")
        guess = int(input("Enter any number "))
    else:
        break
print(" right guess ")


```
